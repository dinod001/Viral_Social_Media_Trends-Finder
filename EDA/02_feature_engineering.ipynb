### 1. Import dependecies

```python
import pandas as pd
import seaborn as sns
```

### 2. Loading Data

```python
df = pd.read_csv('data/processed/Missing_Values_Handled.csv')
df.head()
```

### 3. Creating new features

#### 3.1 Engagement rates 

##### We calculate total_interaction_rate to measure how much viewers engage with a post relative to its reach.


```python
df['Like_Rate'] = (df['Likes'] / df['Views'])
df['Share_Rate'] = (df['Shares'] / df['Views'])
df['Comment_Rate'] = (df['Comments'] / df['Views'])
df.head()
```

### 3.2 dominant_engagement_type

##### We create the dominant_engagement_type feature to identify whether a post gets the most attention through likes, shares, or comments.


```python
#df['dominant_engagement_type'] = df[['Likes', 'Shares', 'Comments']].idxmax(axis=1)
df['dominant_engagement_type'] = df[['Likes','Shares','Comments']].div(df['Views'], axis=0).idxmax(axis=1)
```
